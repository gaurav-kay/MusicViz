## plot windows

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("./build/op", names=["frame", "idx", "real", "imag"])
```

```python
len(df)
```

```python
import math

def mel_scale(f):
    if f < 1000:
        return (3 * f / 200)
    else:
        ret = f / 1000
        ret = math.log(ret) / math.log(6.4)
        ret = 15 + 27 * ret
        return ret
```

```python
df["freq"] = (44100 * df["idx"]) / 4410
df["mag"] = (df["real"] ** 2 + df["imag"] ** 2) ** 0.5
df["mag"] = (df["mag"] - df["mag"].min()) / (df["mag"].max() - df["mag"].min())
df["mel"] = df["freq"].apply(mel_scale)
# df["mel"] = (df["mel"] - df["mel"].min()) / (df["mel"].max() - df["mel"].min())
```

```python
len(df.groupby(by=["frame"]))
```

```python
df[df["frame"] == 0]
```

```python
for frame, frame_df in df.groupby(by=["frame"]):
    plt_df = frame_df[(df["idx"] < (4410 // 2)) & (df["freq"] < 12_000)]
    # plt.plot(plt_df["freq"], plt_df["mag"], color='black')
    plt.plot(plt_df["mel"], plt_df["mag"], color='black')
    plt.ylim(0, 1)
    plt.savefig(f"./frames/{frame[0]}_frame.jpg", dpi=100)
    plt.close()
    
```
